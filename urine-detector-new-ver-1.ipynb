```python
import pandas as pd
import numpy as np
import os

df = pd.read_excel("urine_data.xlsx", sheet_name="AI Training")

# Filter the DataFrame to exclude rows where 'Name' contains "iPhone"
df = df[~df['Name'].str.contains("iPhone")]
```

```python
df = df[df['Position Mark'].str.startswith('F')]
```

```python
# Calculate brightness as the average of R, G, and B
df['brightness'] = df[['R', 'G', 'B']].mean(axis=1)

# Calculate another feature as the sum of R, G, and B
df['rgb_sum'] = df['R'] + df['G'] + df['B']

display(df.head())
```

```python
import cv2

def get_image_brightness(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Warning: Could not load image at {image_path}")
        return None

    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    brightness = np.mean(gray_img)

    return brightness
```

```python
from tqdm import tqdm

df['brightness'] = np.nan

for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    image_name = row['Name']
    # Extract subject ID from the image name (e.g., '065' from '065_Samsung Galaxy S22 Ultra_F2.jpg')
    subject_id = image_name.split('_')[0]
    # Construct the full image path using the base directory, subject ID, and image name
    image_path = os.path.join('AI', subject_id, image_name)
    brightness_value = get_image_brightness(image_path)
    df.loc[index, 'brightness'] = brightness_value

display(df.head())
```

```python
# Calculate the mean and standard deviation of the brightness values
mean_b = df['brightness'].mean()
std_b = df['brightness'].std()

# Normalize the brightness values
df['normalized_brightness'] = (df['brightness'] - mean_b) / std_b

display(df.head())
```

```python
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import torch
import os

class UrineDataset(Dataset):
    def __init__(self, dataframe, img_dir, tabular_features, transform=None):
        self.dataframe = dataframe
        self.img_dir = img_dir
        self.tabular_features = tabular_features
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, self.dataframe.iloc[idx, 0].split('_')[0], self.dataframe.iloc[idx, 0])
        try:
            image = Image.open(img_name).convert('RGB')
        except Exception as e:
            print(f"Error loading image {img_name}: {e}")
            return None, None, None # Indicate failure to load

        # Extract tabular features and ensure they are numeric before converting to tensor
        tabular_data = self.dataframe.iloc[idx][self.tabular_features].values.astype(float)
        tabular_data = torch.tensor(tabular_data, dtype=torch.float)

        # Assuming 'Sp.Refractometer' is your target variable
        label = self.dataframe.iloc[idx, self.dataframe.columns.get_loc('Sp.Refractometer')]
        # Scale the label to the 5-30 range
        scaled_label = (label * 1000) - 1000
        scaled_label = torch.tensor(scaled_label, dtype=torch.float)


        if self.transform:
            image = self.transform(image)

        return image, tabular_data, scaled_label # Return the scaled label

# Define transforms
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)), # Resize images to a consistent size
    transforms.ToTensor(),         # Convert images to PyTorch tensors
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # Normalize pixel values
])

# Create Dataset instance
image_directory = 'AI'
tabular_features = ['R', 'G', 'B', 'brightness', 'rgb_sum', 'normalized_brightness']
full_dataset = UrineDataset(df, image_directory, tabular_features=tabular_features, transform=data_transforms)

# Filter out samples where image loading failed
valid_indices = [i for i in tqdm(range(len(full_dataset)), desc="Filtering dataset") if full_dataset[i][0] is not None]
dataset = torch.utils.data.Subset(full_dataset, valid_indices)


# Split dataset into training and testing (example split)
train_size = int(0.9 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])

# Create DataLoaders
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

print(f"Training dataset size: {len(train_dataset)}")
print(f"Testing dataset size: {len(test_dataset)}")
```

```python
# Define the list of tabular features
tabular_features_lgbm = ['R', 'G', 'B', 'brightness', 'rgb_sum', 'normalized_brightness']

# Extract tabular features and target variable
X_tabular = df[tabular_features_lgbm]
y_target = df['Sp.Refractometer']

# Convert to NumPy arrays
X_tabular_np = X_tabular.to_numpy()
y_target_np = y_target.to_numpy()

print("Tabular features (X_tabular_np) shape:", X_tabular_np.shape)
print("Target variable (y_target_np) shape:", y_target_np.shape)
```

```python
import torch.nn as nn
import torchvision.models as models

class HybridModel(nn.Module):
    def __init__(self, num_tabular_features):
        super(HybridModel, self).__init__()
        # Load a pre-trained ResNet-18 model
        resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

        # Remove the final fully connected layer
        self.image_feature_extractor = nn.Sequential(*list(resnet.children())[:-1])

        # Define layers for tabular features
        self.tabular_processor = nn.Sequential(
            nn.Linear(num_tabular_features, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU()
        )

        # Define the final regression layer
        # The input size is the sum of image features (512 for ResNet-18) and tabular features (32 after processing)
        self.regressor = nn.Linear(512 + 32, 1)

    def forward(self, image, tabular_data):
        # Process image data
        image_features = self.image_feature_extractor(image)
        image_features = torch.flatten(image_features, 1) # Flatten the features

        # Process tabular data
        tabular_features = self.tabular_processor(tabular_data)

        # Concatenate features
        combined_features = torch.cat((image_features, tabular_features), dim=1)

        # Predict the target
        output = self.regressor(combined_features)
        return output

# Instantiate the model
num_tabular_features = len(tabular_features) # Use the length of the defined tabular_features list
model = HybridModel(num_tabular_features)

# state_dict = torch.load('/kaggle/input/25_epochs_urine/pytorch/default/1/25_epochs.pth', weights_only=True)

# model.load_state_dict(state_dict)

# Move model to device if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("HybridModel instantiated and moved to device.")
```

```python
import torch.optim as optim
import torch.nn as nn
from tqdm import tqdm

# Define Loss Function and Optimizer
criterion = nn.MSELoss() # Mean Squared Error for regression
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the Hybrid Model
num_epochs = 50 # You can adjust the number of epochs

# Move model to device if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

for epoch in range(num_epochs):
    model.train() # Set the model to training mode
    running_loss = 0.0
    for images, tabular_data, labels in tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        # Move data to the device
        images = images.to(device)
        tabular_data = tabular_data.to(device)
        labels = labels.to(device).unsqueeze(1) # Add a dimension for the loss calculation

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images, tabular_data)

        # Calculate the loss
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

    epoch_loss = running_loss / len(train_dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")

print("Training finished.")
```

```python
from sklearn.metrics import mean_absolute_error
import torch

# Evaluate the Hybrid Model
model.eval() # Set the model to evaluation mode
test_loss = 0.0
all_labels = []
all_outputs = []

with torch.no_grad(): # Disable gradient calculation during evaluation
    for images, tabular_data, labels in test_dataloader:
        # Move data to the device
        images = images.to(device)
        tabular_data = tabular_data.to(device)
        labels = labels.to(device)

        # Forward pass
        outputs = model(images, tabular_data)

        # Calculate the loss
        loss = criterion(outputs, labels.unsqueeze(1)) # Add a dimension for the loss calculation

        test_loss += loss.item() * images.size(0)

        all_labels.extend(labels.cpu().numpy())
        all_outputs.extend(outputs.cpu().numpy())

# Calculate average test loss
average_test_loss = test_loss / len(test_dataset)
print(f"Test Loss (MSE): {average_test_loss:.4f}")

# Calculate MAE
mae = mean_absolute_error(all_labels, all_outputs)
print(f"Test MAE: {mae:.4f}")

print("Evaluation finished.")
```
